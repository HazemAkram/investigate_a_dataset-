### Creating temp files to customize in  

```python
df_income_clean = df_income.copy()
df_life_clean = df_life.copy()
df_population_clean = df_population.copy()
```

- drop all years after 2020

```python
df_income_clean.drop(df_income_clean.iloc[:,222:], axis='columns', inplace=True)
df_income_clean
```

```python
df_life_clean.drop(df_life_clean.iloc[:,222:], axis='columns', inplace=True)
df_life_clean
```

```python
df_population_clean.drop(df_population_clean.iloc[:,222:], axis='columns', inplace=True)
df_population_clean
```

- handel null values in life expectancy table 

```python
countries = ['Andorra','Dominica','Marshall Islands']
for c in countries : 
    df_life_clean.drop(df_life_clean.loc[df_life_clean['country'] == c].index, inplace=True)
```

```python
df_life_clean.isnull().sum().sum()
```

```python
df_life_clean
```

- drop missing countries in population & incom tables 

```python
df_diff1 = pd.concat([df_life_clean['country'],df_population_clean['country']]).drop_duplicates(keep=False)
df_diff1
```

```python
df_diff2 = pd.concat([df_life_clean['country'], df_income_clean['country']]).drop_duplicates(keep=False)
df_diff2
```

```python
countries = ['Andorra','Dominica','Holy See','Liechtenstein','Marshall Islands','Monaco','Nauru','Palau','San Marino','St. Kitts and Nevis','Tuvalu']
for c in countries : 
    df_population_clean.drop(df_population_clean.loc[df_population_clean['country'] == c].index, inplace=True)
    df_income_clean.drop(df_income_clean.loc[df_income_clean['country'] == c].index, inplace=True)
    df_life_clean.drop(df_life_clean.loc[df_life_clean['country'] == c].index, inplace=True)
```

```python
df_population_clean.shape
```

```python
df_income_clean.shape
```

```python
df_life_clean.shape
```

- melting year columns into one column 
- multi-index to ('country','year')
- change Year column from string to datetime

```python
# life_expectancy DataFrame
df_life_clean= pd.melt(df_life_clean ,id_vars=['country'])
df_life_clean[df_life_clean['variable'] == '1800']
```

```python
df_life_clean.rename(columns = {'variable': 'Year', 'value': 'life_expectancy'}, inplace = True)
df_life_clean.sort_values(by=['country','Year'], inplace=True)
df_life_clean
```

```python
df_life_clean['Year'] = pd.to_datetime(df_life_clean['Year'])
df_life_clean['Year'] = df_life_clean['Year'].dt.year
df_life_clean.info()
```

```python
df_life_clean.set_index(['country','Year'], inplace=True)
df_life_clean
```

```python
#income_per_person DataFram
df_income_clean = pd.melt(df_income_clean ,id_vars='country')
df_income_clean[df_income_clean['variable'] == '1800']
```

```python
df_income_clean.rename(columns={'variable':'Year', 'value':'GDP/capita'}, inplace=True)
df_income_clean.sort_values(by=['country','Year'], inplace=True)
df_income_clean
```

```python
df_income_clean['Year'] = pd.to_datetime(df_income_clean['Year'])
df_income_clean['Year'] = df_income_clean['Year'].dt.year
df_income_clean.info()
```

```python
df_income_clean.set_index(['country','Year'], inplace=True)
df_income_clean
```

```python
# population DataFrame
df_population_clean = pd.melt(df_population_clean, id_vars='country')
df_population_clean[df_population_clean['country'] == 'Afghanistan']
```

```python
df_population_clean.rename(columns={'variable':'Year','value':'population'}, inplace=True)
df_population_clean.sort_values(by=['country','Year'], inplace=True)
df_population_clean
```

```python
df_population_clean['Year'] = pd.to_datetime(df_population_clean['Year'])
df_population_clean['Year'] = df_population_clean['Year'].dt.year
df_population_clean.info()
```

```python
df_population_clean.set_index(['country','Year'], inplace=True)
df_population_clean
```

- merge 3 Datasets into one Dataset

```python
df_clean = df_population_clean.merge(df_income_clean, how='inner', on=['country','Year'])
gapminder_clean = df_clean.merge(df_life_clean, how='inner', on=['country','Year'])
gapminder_clean
```

- add GDP_total & GDP_groth columns 
    - GDP = GDP/capita * popualtion 


```python
gapminder_clean['GDP'] = gapminder_clean['population'] * gapminder_clean['GDP/capita']
gapminder_clean
```

```python
gapminder_clean.loc['Egypt'].loc[1800]
```

```python
# save to master file.csv
gapminder_clean.to_csv('Gapminder_master.csv',  encoding='utf-8')
```

## Exploratory Data Analysis

```python
# custom function to plot data 
def custom_plot(x) :
    plt.plot(gapminder_clean.loc[x]['life_expectancy'],label = x)
    plt.title(x + ' life expectancy')
    plt.xlabel('Year')
    plt.ylabel('life expectancy')
    plt.legend()
    plt.show();
```

#### Sweden
> Life expectancy in Sweden was 32.16 in the year 1800, and over the course of the next 221 years, it is expected to have increased to 82.6 by 2020. Although life expectancy has generally increased throughout Sweden's history, there was a lot of fluctuation around the turn of the nineteenth century due to The Napoleonic Wars and First Cholera Epidemic, **and again in the 1910s due to the Spanish Flu Epidemic.**

```python
custom_plot('Sweden')
```

#### Finland
> Life expectancy in Finland was 36.6 in the year 1800, and over the course of the next 221 years, it is expected to have increased by over double this, to 81.9 years by 2020. Although life expectancy has generally increased throughout Finland's history, it fluctuated greatly between 1800 and 1875, in the 1910s because of the Spanish Flu epidemic that swept across the world and the Finnish Civil War, and again during the Second World War in the 1930s and 40s.

```python
custom_plot('Finland')
```

#### Spain
> Life expectancy in Spain was below thirty in the year 1800, it is expected to have increased to 83.5 by the year 2020. Although life expectancy has generally increased throughout Spain's history, there were several times where the rate deviated from its previous trajectory. The most notable changes were a result of the global influenza epidemic of the 1910s, (called the 'Spanish Flu' as it received more press coverage in neutral Spain, during the First World War) and the Spanish Civil War two decades later.

```python
custom_plot('Spain')
```

#### ام الدنيا
> Between 1900 and 1950, Egypt’s life expectancy saw little change, hovering around 33 years, with the only decrease coming in the 1910s during the Spanish Flu epidemic and the Second World War. However, following the removal of the monarchy and establishment of the Egyptian republic in the 1952 Revolution, life expectancy saw a several decade rise, largely due to a significant expansion of the economy, health services, and other forms of social welfare; this increase then slowed between 1960 and 1975 as the economic growth in the country slowed, but it then increased to 68 years at the turn of the millennium. Since 2000, this growth has slowed, as a mixture of high population growth, high unemployment, and significant civil unrest have hit the country in the last two decades, although it has still reached 72 years in 2020, which is just one year below the global life expectancy.

```python
gapminder_clean.loc['Egypt'][['life_expectancy','GDP/capita','GDP']].plot(subplots=True, layout=(2, 3), figsize=(15, 6), sharex=False, title = 'Egypt life expectancy, GDP/capita & GDP');
```
